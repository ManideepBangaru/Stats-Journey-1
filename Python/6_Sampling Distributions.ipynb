#### Degrees of Freedom
It means the number of items to be selected freely out of n items. It is **(n-1)**. It is denoted by **df**.

For example, 40+50+10 = 100 \
One can choose only two items freely, and the selection of third value cannot be made freely. If you select 40 and 50, then the third value should be 50.\

degrees of freedom = n - 1 , here n = 3 \
hence, 3 - 1 = 2

Therefore, degrees of freedom for this particular case is 2

### Random Errors / Sampling Errors
Sometimes, sample may not provided the exact inferences about the population. It is because a sample is a portion of the population. Such type of errors are referred to as **Random Errors** or **Sampling Errors**

### Sampling distribution of the Mean
It is the distribution of all the possible sample means if you select all possible samples of a given size

#### The Unbiased property of the Sample Mean
The Sample mean is **unbiased** because the mean of all the possible sample means (of a given sample size, n) is equal to the population mean ($\large \mu$).

When you have data from a population, you compute the population mean and standard deviations by using below equations :

#### Population Mean

$\large \mu = \Large \frac{\sum_{i=1}^{N} X_i}{N}$

#### Population Standard deviation

$\large \sigma = \sqrt{\Large \frac{\sum_{i=1}^{N} (X_i-\mu)^2}{N}}$

### Standard error of the Mean

+ The Sample mean varies from sample to sample
+ Sample means are less variable than that of individual values in the population
+ Each sample mean averages together all the values in the sample
+ A population consists of individual outcomes that can take on a wide range of values, from extremely small to extremely large
+ However, if a sample has extremely large value, although this value will have an effect on the sample mean, the effect is reduced because the value is averaged with all the other values in the sample
+ As the sample size increases, the effect of a single extreme value becomes smaller because it is averaged with mnore values

The value of the standard deviation of all possible sample means, called the **Standard error of the mean**. And it expresses how the sample means vary from sample to sample.

$\Large \sigma_{\bar{X}} = \Large \frac{\Large \sigma}{\Large \sqrt{n}}$

To find Z for the Sampling distribution of the Mean

$ \Large Z = \Large \frac{\bar{X}-\mu}{\Large \frac{\Large \sigma}{\Large \sqrt{n}}}$

### Sampling from Non-normally Distributed Populations - The Central Limit Theorem

However, for many analyses, you will either be able to know that the population is not normally distributed or conclude that it would be unrealistic to assume that the population is normally distributed. An important theorem in statistics, the **Central Limit Theorem**, deals with these situations.

#### The Central Limit Theorem
As the sample size (the number of values in each sample) gets large enough, the sampling distribution of the mean is approximately normally distributed. This is true regardless of the shape of the distribution of the individual values in the population.

If you are thinking about what sample size is large enough? As a general rule, statisticians have found that for many population distributions, when the sample size is at least 30, the sampling distribution of the mean is approximately normal.

However, you can apply central limit theorem for even smaller sample sizes if the population distribution is approximately bell-shaped.
