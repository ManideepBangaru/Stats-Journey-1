$\sf \normalsize \textbf{Nature of Statistical law : }$

**Law of Statistical regularity** states that from a very large population, if you select a moderate large sample, then the sample is highly likely to have the characteristics of the population.

**Law of inertia of large numbers** is by principles of law of statistical regularity states that if the sample is bigger in size, then the study yields the results that of population.

<font color = 'olive'>$\sf \Large \textbf{Collecting data}$</font>

### Population and Samples
* **Population :** A population contains all the items or individuals of interest that you seek to study
* **Sample :** A sample contains only a proportion of a population of interest

### Data Sources
* Capturing data generated by ongoing business activities
* Distributing data compiled by an organization or individual
* Compiling the responses from a survey
* Conducting a designed experiment and recording the outcomes of the experiment
* Conducting an observational study and recording the results of the study

#### Primary data sources
When you perform the activity that collects the data, you are using a primary data source

#### Secondary data sources
When the data collection part is done by someone else, you are using a secondary data source

#### Internal Source
Set of internal data if it is obtainable from well within an organization and relates to organization‚Äôs operations
Ex:- Computerized files containing sales figures, financial data,operating information and so on

#### External Source
Set of data that is collected outside the organization. This type of data may be available in the published financial periodicals or it may be stored in an internal computer data bank accessible  by an on-line computer terminal


### Data Structures

* **Univariate Data** : Univariate data sets have a single piece of information recorded for each item.
**Ex:-** Statistical analysis of data collected regarding the income level through a marketing survey would reveal the distribution of incomes, specific income level, variation in the income level and the number of people within any given range of income

* **Bivariate Data** : Bivariate (two variables) data sets have exactly two pieces of information recorded for every item. Application of statistical analysis would reveal the relationship between the two variables under study.

* **Multivariate Data** : Multivariate data (Minimum of three variables) sets have a minimum of three pieces of information recorded for every item. Statistical analysis can be applied to study the interrelationship between all the variables.


### Data Collection methods

* **Observation method** (difficult to produce accurate data) : The person who collects the data asks no questions, but he observes carefully the phenomenon under consideration, and records the essential data.

* **Questionnaire method** (more accurate and reliable) : The researcher designs a questionnaire which contains all the relevant questions needed for the study.

* **Mail questionnaire method** (costly one) : The questionnaire is mailed to the respondent‚Äôs residential or official address.

* **Telephonic interview method** (less expensive, somewhat accurate) : The researcher asks the relevant questions to the respondents over the telephone.

* **Census method** : When the researcher collects the data from each and every individual of the population
 * **Adv -** complete information, accurate
 * **Disadv -** lot of time and huge money

* **Sampling methods** : When you collect data by selecting a sample, you begin by defining the frame. The frame is a complete or partial listing of the items that make up the population from which the sample will be selected.

### Types of Sampling methods

#### Non probability sample
* Judgement Sample - Collect opinions of Subject matter experts (cannot generalize to population)
* Convenience Sample - Select the items that are inexpensive and convenient to take (ex:- picking a book which is on top most stack)

#### Probability sample
* Simple Random Sample
* Systematic Sample
* Stratified Sample
* Cluster Sample

* Simple Random Sample : Most used, but it has its own disadvantages, subjected to more variation than other sampling methods. When the population is very large, carrying out a simple random sample may be time consuming and expensive
    * Sample with replacement : Probability of picking a item from population N is 1/N, and for the next item 1/N as the previously picked item is replaced in the population
    * Sample without replacement : Probability of picking a item from population N is 1/N, and for the next item 1/(N-1) as the previously picked item is not replaced in the population
    
> <font color='red'> _Sample without replacement is usually preferred_ </font>

* Systematic Sample : 
    * You partition the N items into n groups of k items, where k=(N/n) , you round k to the nearest integer.
    * You choose the first item to be selected at random from the first k items in the population. 
    * Then you select the remaining n-1 items by taking every kth item thereafter from the entire population.
    
> <font color = 'red'> _Simple random and Systematic sampling are simpler than any other sampling methods. But they require more data.In addition, Systematic is prone to Bias_ </font>

* Stratified Sample :
    * To overcome inefficiency of simple random and potential selection bias involved in systematic, Stratified and Cluster sampling comes into picture.
    * You first subdivide the N items in the population into separate subpopulations, or strata. 
    * A stratum is defined by some common characteristic, such as gender or year in school. 
    * You select a simple random sample within each for the strata and combine the results from the separate simple random samples
    
> <font color = 'red'> _Expensive to implement_ </font>

* Cluster Sample :
    * You divide the N items in the population into clusters that contain several items
    * Clusters are often naturally occurring groups such as counties, election districts, city blocks, household, or sale territories

> <font color = 'red'> _Requires a large sample size_ </font>

### Data Preparation

$\sf \large \textbf{Excercises :}$

1) Create a simple random sample with replacement of size 40 from a population of 800 items

2) Create a simple random sample without replacement of size 40 from a population of 800 items

```python
import numpy as np
import random
random.seed(123)
elements = [i for i in range(1,800)]
print('With replacement    = ',str(len(np.unique(random.choices(elements,k = 40)))))
print('Without replacement = ',str(len(np.unique(random.sample(elements,k = 40)))))
```

### Organizing Numerical variables

* An ordered array arranges the data for a numerical variable in rank order, from smallest to the largest value. It helps you get a better sense of range of values especially when you have more than a few values

* **The Frequency distribution**
    * It tallies the values of a numerical variable into a set of numerically ordered classes. 
    * Each class groups a mutually exclusive range of values, called as a class interval
    * To create a useful frequency distribution, you must consider how many classes would be appropriate for your data as well as determine a suitable width for each class interval
    * In general, a frequency distribution should have at least 5 and no more than 15 classes because having too few or too many classes provides little new information
    * To determine the class interval width, you subtract the lowest value from the highest value and divide that result by the number of classes you want the frequency distribution to have

        <font color='blue'>Interval Width = $\large \frac{\text{highest value - lowest value}}{\text{number of classes}}$</font>

* **Frequency distribution looks :**
![](https://d1whtlypfis84e.cloudfront.net/guides/wp-content/uploads/2018/02/17080119/table.jpg)

* **The Relative frequency and Percentage distribution**
    * It presents the relative frequency, or proportion of the total for each group that each class represents 
    * A percentage distribution, presents the percentage of the total for each group that each class represents
    * The proportion or relative frequency, in each group is equal to the number of values in each class divided by total number of values
    * The percentage in each group is its proportion multiplied by 100%

* **The Relative frequency and Percentage distribution looks :**
![](https://www.softschools.com/math/probability_and_statistics/images/frequency_table_categorical_data_2.png)

* **The Cummulative distribution**

    It provides a way of presenting the information about the percentage of values that are less than a specific amount

    ![](https://www.statisticshowto.com/wp-content/uploads/2012/11/cum-freq.jpg)

* **Types of Class Interval**
    * **Exclusive method :** In this, class intervals are arranged in such a way that upper limit of one class is the lower limit of the subsequent class
    * **Inclusive method :** Both the lower and the upper limits are included in the same interval
    * **Open-end method :** 
        * The lower limit of the initial class and the upper limit of the upper class will not be given.
        * This is possible when there is a large gap between the minimum and maximum values

### Visualizing Categorical Variables
* Visualizing categorical variables involves making choices about how you seek to present your data
* To highlight how categories directly compare to each other, you use a **Bar chart**
* To highlight how categories form parts of a whole, you use a pie or **doughnut chart**
* To present data that are concentrated in only a few of your categories, you use a **Pareto chart**
* To visualize two categorical variables together, again thinking about what you want to highlight
    * To highlight direct comparisons, you use a **side-by-side chart**
    * To highlight how parts from a whole, you use a **doughnut chart**

#### Bar chart
* A bar chart visualizes a categorical variable as a series of bars, with each bar representing the tallies for a single category. 
* In a bar chart the length of each bar represents either the **frequency** or **percentage** of values for a category and each bar is separated by a space
